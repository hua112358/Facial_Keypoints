```python
from data_load import FacialKeypointsDataset, Resize, Crop, Normalize, ToTensor
import torchvision
import torch
from models import Net
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
"""
    Step 0: Get Data

"""

transform = torchvision.transforms.Compose([Resize(250), Crop(224), Normalize(), ToTensor()])
train_dataset = FacialKeypointsDataset(csv_file="data/training_frames_keypoints.csv", 
                                       root_dir="data/training/", 
                                       transform=transform)
train_loader = torch.utils.data.DataLoader(train_dataset, 64, True, num_workers=4)
test_dataset = FacialKeypointsDataset(csv_file="data/test_frames_keypoints.csv", 
                                       root_dir="data/test/", 
                                       transform=transform)
# test_loader = torch.utils.data.DataLoader(test_dataset, 64, True, num_workers=4)
test_loader = torch.utils.data.DataLoader(test_dataset, 20, True, num_workers=4)
```

```python
"""
    Step 1: Load Model
    
"""

# # load your own model 
# net = Net()
# print(sum([param.numel() for param in net.parameters()]))
# print(net)

# # device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# # net = net.to(device)

# if torch.cuda.is_available():
#     net = net.cuda()
    

# # tranfering learning, load pytorch model: resnet18, alexnet, ...

# # 1 transfering as finetune net
# net = torchvision.models.resnet18(pretrained=True)
# net.conv1 = torch.nn.Conv2d(1, 64, 7, 2, 3, bias = False)
# net.fc = torch.nn.Linear(net.fc.in_features, 136)
# torch.cuda.empty_cache()
# print(sum([param.numel() for param in net.parameters()]))
# print(net)
# if torch.cuda.is_available():
#     net = net.cuda()
# optimizer = torch.optim.Adam(net.parameters(), lr = 0.001, weight_decay=5e-4)

# 2 transfering as feature extractor
net = torchvision.models.resnet18(pretrained = True)
net.conv1 = torch.nn.Conv2d(1, 64, 7, 2, 3, bias = False)
for param in net.parameters():
    param.requires_grad = False
list(net.parameters())[0].requires_grad = True
list(net.parameters())[1].requires_grad = True
net.fc = torch.nn.Linear(net.fc.in_features, 136)
torch.cuda.empty_cache()
print(sum([param.numel() for param in net.parameters()]))
print([param.numel() for param in net.parameters()])
print(net)
if torch.cuda.is_available():
    net = net.cuda()
optimizer = torch.optim.Adam(net.parameters(), lr = 0.001, weight_decay=5e-4)
```

```python
%%time

"""
    Step 2: Training 

"""

criterion = torch.nn.SmoothL1Loss()
# optimizer = torch.optim.SGD(net.parameters(), lr = 0.001, weight_decay=5e-4)
optimizer = torch.optim.Adam(net.parameters(), lr = 0.001, weight_decay=5e-4)

for epoch in range(20):
    print("epoch " + str(epoch + 1))
    if (epoch + 1) % 20 == 0:
        for p in optimizer.param_groups:
            p["lr"] *= 0.1
    net.train()
    running_loss = 0.0
    for batch_index, data in enumerate(train_loader):
        print("batch_index " + str(batch_index + 1))
        image, keypoints = data["image"], data["keypoints"]
        image, keypoints = image.type(torch.FloatTensor), keypoints.type(torch.FloatTensor)
        if torch.cuda.is_available():
            image, keypoints = image.cuda(), keypoints.cuda()
        keypoints = keypoints.view(keypoints.shape[0], -1)
        pred_keypoints = net(image)
        loss = criterion(pred_keypoints, keypoints)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        if batch_index % 50 == 49:
            print("epoch " + str(epoch + 1) + " batch_index " + str(batch_index) + " loss " + str(running_loss / (50)))
            running_loss = 0.0
    
    eval_loss = 0.0
    net.eval()
    for data in test_loader:
        image, keypoints = data["image"], data["keypoints"]
        image, keypoints = image.type(torch.FloatTensor), keypoints.type(torch.FloatTensor)
        keypoints = keypoints.view(keypoints.shape[0], -1)
        if torch.cuda.is_available():
            image, keypoints = image.cuda(), keypoints.cuda()
        with torch.no_grad():
            pred_keypoints = net(image)
        loss = criterion(pred_keypoints, keypoints)
        eval_loss += loss.item()
    print("Eval Loss " + str(eval_loss / len(test_loader)))
print("Training Finished!")
```

```python
"""
    Step 3: Evaluate Model

"""

test_image = test_dataset[0]["image"]
test_keypoints = test_dataset[0]["keypoints"] * 50 + 100

pred_keypoints = net(test_image.type(torch.FloatTensor).reshape((1, 1, 224, 224)).cuda())
pred_keypoints = pred_keypoints.data.cpu().numpy().reshape((68, 2)) * 50 + 100

plt.figure()
plt.imshow(test_image.data.numpy().squeeze() * 255)
plt.scatter(test_keypoints[:, 0], test_keypoints[:, 1], s = 20, marker = ".", c = "r")
plt.scatter(pred_keypoints[:, 0], pred_keypoints[:, 1], s = 20, marker = ".", c = "m")

conv1_weight = net.conv1.weight.data.cpu().numpy()
filter_index = 0
conv1_filter = conv1_weight[filter_index][0]
filter_image = cv2.filter2D(test_image.data.numpy().squeeze() * 255, -1, conv1_filter)
plt.figure()
plt.imshow(filter_image, cmap="gray")
```

```python
"""
    Step 4: Save and load model

"""
# # save and load entire model
# torch.save(net, model_dir + model_name)
# net = torch.load(model_dir + model_name)

# save and load weights
model_dir = "saved_models/"
model_name = "facial_keypoints_model_1.pt"
torch.save(net.state_dict(), model_dir + model_name)

net = Net()
net.load_state_dict(torch.load(model_dir + model_name))
net.eval()
```

### Question 1: What optimization and loss functions did you choose and why?


**Answer**: I choose Adam as my optimzier and SmoothL1loss as my loss function.

First, as the loss function, this is a regression problem and I need to use a regression loss, such as L1 loss or MSE loss. If we use MSE loss, when the error is large, the gradient will contain |x-t|, x means predict coordinate and t means ground truth, then the gradient will explod, so we want graident more stable when error is large. In this thought, I use SmoothL1Loss.

As for optimizer, Adam can converge faster than SGD, so we choose Adam.

### Question 2: What kind of network architecture did you start with and how did it change as you tried different architectures? Did you decide to add more convolutional layers or any layers to avoid overfitting the data?

**Answer**: I started a very simple network architecture just like LeNet. But I found it's not very effective becuase of the shallow layers. I just change the architecture to the most popular architecture, resnet. I use resnet-18 as the final architecture trading off the speed and accuracy. In order to avoid overfitting, I add weight decay in optimzier which is like regularization.

### Question 3: How did you decide on the number of epochs and batch_size to train your model?

**Answer**: I set 64 as the batch size becasue too small batch will lead to batchnorm unstable and too large batch will out of memory. I just print test loss to check if overfitting, this decides the number of epochs after many experiments.
